```python
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percentual_bonus = 0.1
bonus_extra = 50
for vendas in lista_vendas:
    bonus = 0
    if vendas >= meta:
        bonus = vendas * percentual_bonus
        if venda >= 2000:
            bonus += bonus_extra
        print(f"vendas: {vendas}-> bonus: R${bonus}")
```
